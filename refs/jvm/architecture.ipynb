Based on Oracle JDK 13.  
You can find the JVM speficiation [here](https://docs.oracle.com/javase/specs/)

The JVM is a vartual machine that executes bytecode.  
It is stack-based (no registers).

# Structure

Compiled code is represented into a .class file, with a platform-independant format.  

It has primitive and reference types.
Instructions operate on some specific type (eg addi for int, addf for float), but the bytecode isn't typechecked.  
Objects are of type reference, they are pointers.

## Data types

- byte: 8-bit signed
- short: 16-bit signed
- int: 32-bit signed
- long: 64-bit signed
- char: 16 bit unsigned, UTF-16.
- float: 32bit fp
- double: 64bit fp
- boolean: true/false. Actually bool are usualy manipulated as int
- returnAddress: pointer of JVM instructions

When manipulating byte, short, char, they are internally widened to int by JVM

## Reference types

- class type: reference to class instance
- array type: reference to array, contains component type (may also be array)
- inteferace: reference to interface instance

## Run-time data areas

### PC

Each JVM thread has its own PC (Program Counter)

### Stack

Each JVM thread has its own stack. It store function frames.

### Heap

Shared among all JVM threads.  
Data not directly deallocated, use of Garbage Collector.

### Method Area

Shared among all JVM threads. Similar to text segment. Per-class structures with:
- runtime constant pool: compile-time known literals, method and field reference (for dynamic dispatch)
- field and method data
- method bytecode

### Native method stack

Each JVM thread has its own native (C) stack, to call native code.

## Frame

Part of JVM stack, frame created at each method call. Contains:
- local variables (and function arguments)
- operands stack 
- reference to runtime constant pool of the current class.

### Local variable

Array of local variables, of size known at compile time.  
For long and double type, need 2 local variables.  
Accesed by array indexing. For long/double, index n to access value in (n, n+1).  
Method arguments are stored at beginning of array. For instance method, variable 0 contains reference to object.

### Operand stack

LIFO operand stack. Size known at compile time for each frame.  
Instructions usually pops data from operand stack, operates on them, then push result to operand stack.  
Also used to store parameters for method call, and return value of method call.  
Long / double takes 2 entries in operand stack.

### Dynamic Linking

Reference to runtime constant pool.  
Methods and class fields are refered using symbolic references. 
Dynamic linking transform this references into exact offset for fields, and method adresses.


## Special Methods

### Instance initialization methods

Void function, with special name $<$init$>$. Corresponds to class constructors.

### Class initialization methods

Void function, with special name $<$clinit$>$. Used to init static fields of class.

### Signature polymorphic methods

Related to java.lang.invoke. Allows to control low-level dynamic call resolution in Java.

## Exception

Instance of class Throwable, result in immediate nonlocal transfer or control.  

JVM throws when:
- run athrow instruction
- abnormal execution of an instructiom
- asynchronous exception: stop method of thread called, or internal JVM error.

## Instructions

An instructor is 1 op-code byte, followed by 0 or more operands.  
The instruction encode type information about it's operands (eg: iadd to add int, ifloat to add floats, they have 2 different opcodes).
