# CLI

CLI (Common Language Infrastructure) is an open stardind describing a runtime environment to allow for multiple languages to be used, independant on the platform and the hardware, and to communicate between each other.

It consists or 4 parts:

- The Common Type System (CTS): A set of basic datatypes. All compliant languages are able to match their datatypes to these.
- The Metadata: Information about the program structure (class, fields, methods, arguments types), inpedendant of the language. These are generated by the compiler and stored in the assembly.
- The Common Language Specification (CLS): A set of rules every language respect. This allows communication between languages (call of methods, exceptions handling).
- Virtual Execution System (VES): Responsible for loading and executing programs.

All compatible languages compile to the same assembly, which is platform and architecture independant: CIL

# CIL (Common Intermediate Language)

CIL is stack-based and object-oriented.  
There as instructions to create objects, access fields, call methods, and do dynamic dispatch.  
Actually it is really similar to the Java bytecode, both have mostly the same features.  

A difference with the Java bytecode is that they are many instructions to handle pointer and memory operations, and call with function pointers, that are not available in Java bytecode. This can allow to compile C/C++ to CIL.

During execution, The CIL is usually compiled to native code with JIT. OAT compilation is also possible, losing executable portability.


# CLR

CLR (Common Language Runtime) execution engine.
This is an implementation of the CLI,
Provide environment for managed executables, for .NET languages, such as C\#, VB.NET, F\#.

Has multiple functions: 
- garbage collection
- cross-langage interopability
- exceptions handling
- security
- thread execution
- verification
- optimizations

For example, can call method instance of a class from one language to another, or handle exceptions thrown by another language.
