# Execution Provider

An execution provider is an abstraction of a runtime implementation.  
When executing a graph, it's divided in subraphs, for each execution provider depending on its capabilities.

An execution provided has multiple components:

- multiple memory allocators

- data-transfer object: to copy from / to device

- compute capabilities (which kind of nodes in can execute)

- kernel registry: one per provider. Used to register kernel definition and kernel creation function.

- execution handle: opaque, used by the specific kernel implementations. Depends on execution provided type

- type: CPU, CUDA, etc


Execution providers have operations implementations, called kernels.  
There is a default execution provider, for completness, that cover all operations implementations.  
It's considered last when doing the partitioning.
